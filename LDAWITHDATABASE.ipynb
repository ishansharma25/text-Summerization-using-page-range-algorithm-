```python
import pandas as pd
data=pd.read_csv('e:\data\wikihowAll.csv')
data_text=pd.DataFrame(data,columns=['text'])
data_headline=pd.DataFrame(data,columns=['headline'])
def limit(data):
    return(data.head(20))
data_text=limit(data_text.copy())
data_headline=limit(data_headline.copy())
print(data_text)
```

```python
# Create a list of lists from the 'text' column, handling possible float values
text_list = [str(text).split() if isinstance(text, str) else [] for text in data_text['text']]

# Print the list of lists
print(text_list)
```

```python
import string
def remove_punctuation(word):
    return word.translate(str.maketrans('', '', string.punctuation))

text_list_no_punctuation = [[remove_punctuation(word) for word in words] for words in text_list]
print(text_list_no_punctuation)
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Ensure NLTK data is downloaded
nltk.download('punkt')

# Tokenize the text in the list of lists
text_list_tokenized = [word_tokenize(" ".join(words)) for words in text_list_no_punctuation]

# Print the tokenized text
print(text_list_tokenized)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Ensure NLTK data is downloaded
nltk.download('punkt')
nltk.download('stopwords')

# Define a set of English stopwords
stop_words = set(stopwords.words('english'))

# Tokenize the text in the list of lists
text_list_tokenized = [word_tokenize(" ".join(words)) for words in text_list_no_punctuation]

# Remove stopwords from the tokenized text
text_list_no_stopwords = [[word for word in words if word.lower() not in stop_words] for words in text_list_tokenized]

# Print the text with stopwords removed
for words in text_list_no_stopwords:
    print(words)
```
