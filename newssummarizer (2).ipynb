```python

import nltk
def tokenize(text):
    tokens = nltk.word_tokenize(text)
    return tokens
tokens = text.apply(lambda x: tokenize(x))
print(tokens)
```

```python
import nltk
#nltk.download('stopwords')
from nltk.corpus import *
stopword=nltk.corpus.stopwords.words('english')
def removal(text):
  return [word for word in text if word not in stopword]
data_text=tokens.apply(lambda x:removal(x))
print(data_text)
```

```python
from nltk.stem import WordNetLemmatizer
#nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
data_lemmatize=data_text.apply(lambda x:lemmatizer(x))
print(data_lemmatize)
```

```python
from collections import Counter
from collections import defaultdict

def word_frequency(text):
  word_counts = Counter(text)
  return word_counts
word_freq_dict = defaultdict(int)
for text in data_lemmatize:
  for word in text:
    word_freq_dict[word] += 1
word_freq = dict(word_freq_dict)
print(word_freq)
print(type(word_freq))

```

```python
def normalize_word_frequencies(word_freq_dict):
  normalized_word_freq_dict = {}
  max_freq = max(word_freq_dict.values())
  for word, freq in word_freq_dict.items():
    normalized_word_freq_dict[word] = freq / max_freq
  return normalized_word_freq_dict
word_freq_dict = normalize_word_frequencies(word_freq)
print(word_freq_dict)
```

```python

sent_scores = []

for i, sentence in enumerate(data_lemmatize):
    score = 0
    for word in sentence:
        score += word_freq_dict.get(word, 0)
    sent_scores.append(score)
print(sent_scores)
```

```python
select_percentage = 0.3
select_len = int(len(data_lemmatize) * select_percentage)
top_sentence_indices = list(range(len(sent_scores)))
top_sentence_indices = top_sentence_indices[:select_len]
summary = []
for i in top_sentence_indices:
    summary.append(data['text'][i])
print(summary)
```
